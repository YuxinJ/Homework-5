```python
import pandas as pd
violation = pd.read_csv("C:\Users\JYX\Downloads\Building_Violations.csv")
```

```python
#1b
violation_text = ""
for i in range(0,len(violation['VIOLATION INSPECTOR COMMENTS'])):
    if violation['VIOLATION INSPECTOR COMMENTS'][i] != 'NaN':
        violation_text = violation_text + str(violation['VIOLATION INSPECTOR COMMENTS'][i])
```

```python
txtfile = open('newfile.txt','w')
txtfile.write(violation_text)
txtfile.close()
```

```python
#get the lower case 
violation_lower = violation_text.lower()
```

```python
#1c
import string
violation_nonpunct = violation_lower.translate(None, string.punctuation)
```

```python
#1d
import nltk
tokens = nltk.word_tokenize(violation_nonpunct)
```

```python
#1e
from collections import Counter
from nltk.corpus import stopwords
filtered_tokens = [w for w in tokens if not w in stopwords.words('english')]
```

```python
#1f
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
```

```python
#1g
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens]
Counter(stemmed_tokens)
#the most common stem tokens is 'elev'
```
